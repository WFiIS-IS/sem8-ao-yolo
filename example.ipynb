```python
from ultralytics import YOLO

# model = YOLO('yolov8s.pt').cuda()
model = YOLO('yolov8n.pt').cuda()
```

```python
import cv2
import matplotlib.pyplot as plt

# images
img_path = "./zidane.jpg"  # or file, Path, PIL, OpenCV, numpy, list

# read image with OpenCV
img = cv2.imread(img_path)

# convert to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Inference
results = model(img)

# Results
print(results)

annotated_img = results[0].plot()

# plot annotated image
plt.imshow(annotated_img)
```

### Webcam

```python
# show video of webcam

import cv2

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    results = model(frame)
    annotated_img = results[0].plot()
    cv2.imshow('frame', annotated_img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()

cv2.destroyAllWindows()
```

```python
# show video of webcam with fps

import cv2 
import time 

cap = cv2.VideoCapture(0) 
  
prev_frame_time = 0
new_frame_time = 0
  
while(cap.isOpened()): 
  
    ret, frame = cap.read()
    results = model(frame)
    annotated_img = results[0].plot()
  
    new_frame_time = time.time() 
    fps = 1/(new_frame_time-prev_frame_time) 
    prev_frame_time = new_frame_time 
    fps = str(int(fps)) 
  
    cv2.putText(annotated_img, fps, (7, 70), cv2.FONT_HERSHEY_SIMPLEX, 3, (100, 255, 0), 3, cv2.LINE_AA) 
  
    cv2.imshow('frame', annotated_img) 
  
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break
  
cap.release() 
cv2.destroyAllWindows() 
```

### Tracking

```python
# show video of webcam with tracking of objects

import cv2

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    results = model.track(frame, persist=True)
    annotated_img = results[0].plot()
    cv2.imshow('frame', annotated_img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
from collections import defaultdict
import cv2
import numpy as np
import time

cap = cv2.VideoCapture('./vid2.mov')

# Store the track history
track_history = defaultdict(lambda: [])

prev_frame_time = 0
new_frame_time = 0

while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    
    frame = cv2.resize(frame, (640, 480))

    results = model.track(frame, persist=True)

    boxes = results[0].boxes.xywh.cpu()
    track_ids = results[0].boxes.id.int().cpu().tolist()

    annotated_frame = results[0].plot()

    # Plot the tracks
    for box, track_id in zip(boxes, track_ids):
        x, y, w, h = box
        track = track_history[track_id]
        track.append((float(x), float(y)))  # x, y center point
        if len(track) > 30:  # retain 90 tracks for 90 frames
            track.pop(0)

        # Draw the tracking lines
        points = np.hstack(track).astype(np.int32).reshape((-1, 1, 2))
        cv2.polylines(annotated_frame, [points], isClosed=False, color=(100, 255, 0), thickness=10)

    new_frame_time = time.time() 
    fps = 1/(new_frame_time-prev_frame_time) 
    prev_frame_time = new_frame_time 
    fps = str(int(fps)) 

    cv2.putText(annotated_frame, fps, (7, 70), cv2.FONT_HERSHEY_SIMPLEX, 3, (100, 255, 0), 3, cv2.LINE_AA) 

    cv2.imshow("YOLOv8 Tracking", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break
 
cap.release()
cv2.destroyAllWindows()
```
